```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.style as style
style.use(['seaborn-talk', 'seaborn-colorblind'])
plt.rcParams['lines.linewidth'] = 2.5
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams['axes.grid'] = False
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('spacex_launch_data.csv', parse_dates=['Date'])
```

```python
df.head()
```

```python
dates = pd.to_datetime(df['Date'])
payload_time = df.set_index('Orbit')
payload_time[['Payload Mass (kg)']].plot(kind='bar', figsize=(16,9))
```

```python
df.groupby('Orbit').mean().plot(kind='bar', figsize=(16,9))
```

```python
df[['Orbit', 'Flight Number']].groupby('Orbit').count().plot(kind='bar', figsize=(16,9))
```

```python
(
    df[['Booster Version', 'Flight Number']]
    .groupby('Booster Version')
    .count()
    .plot(kind='bar', figsize=(16,9))
)
```
