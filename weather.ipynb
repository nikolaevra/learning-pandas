```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('GlobalLandTemperaturesByMajorCity 2.csv')
```

```python
df['dt'] = pd.to_datetime(df.dt)
df.head()
```

```python
peryear = df.groupby(df.dt.dt.year)[['AverageTemperature']].mean()
# peryear.drop_duplicates()
```

```python
peryear.head()
```

```python
axes = peryear.plot(figsize=(16,9))
axes.set_xlabel('Year')
axes.set_ylabel('Temperature')
```
